```python
def Result_card(Math,English,urdu):
    T_math=100
    T_English=100
    T_Urdu=100
    T_Obt=Math+English+urdu
    T_Marks=T_English+T_math+T_Urdu
    Percentage=(T_Obt/T_Marks)*100
    if(Percentage>=90):
        Grade="A+"
    elif Percentage<=90 and Percentage >=80:
        Grade="A" 
    elif Percentage<=80 and Percentage >=70:
        Grade="B"        
    elif Percentage<=70 and Percentage >=60:
        Grade="C" 
    elif Percentage<=60 and Percentage >=50:
        Grade="D" 
    elif Percentage<=50 and Percentage >=40:
        Grade="E" 
    else:
        Grade="F"
    print("You got :",Grade,"Grade") 

Result_card(50,50,50)
```
